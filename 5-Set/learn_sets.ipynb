```python
myemptyset = set([1,2,3,4,5,6,7,7,8])
print(myemptyset)
```

```python
#basic set operations

#adding and removing element
myset.add(120)
print(myset)
```

```python
#remove element
myset.remove(3)
print(myset)
```

```python
myset.remove(30)
```

```python
myset.discard(23) # if element is not exists in set , it does not give any error okay
#it remove element if it is a member
```

```python
##pop method
removedelement = myset.pop()
print(removedelement)
print(myset)
#first in first out
```

```python
#clear all element
myset.clear()
print(myset)
```

```python
#set membership
myset = {1,2,3,4,5}
print(3 in myset)
print(13 in myset)
```

```python
#mathematical operation
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9,11,12,13}

# Union:

unionset =set1.union(set2)
print(unionset)
```

```python
#intersection
intersection = set1.intersection(set2)
print(intersection)
```

```python
set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

#difference
print(set1.difference(set2)) #set1-set2 , remove common element of both of them
```

```python
print(set2.difference(set1))
```

```python
#symmetric difference
print(set1.symmetric_difference(set2))
#unique elemtnt of all the set combine
```

```python
#sets method
set1 = {1,2,3,4,5}
set2 = {4,5,3}
print(set1.issubset(set2))
```

```python
print(set1.issuperset(set2))
```

```python
lst = [1,2,3,2,3,4,5,6,5,4]
print(set(lst))  #remove all the duplicate element of the list , so simple
```

```python
#counting unique words in text

text = "in this tutorial we are discuss about set"
words  = text.split()

#convert list of word to set to get unique words

uniqueword = set(words)
print(uniqueword)
print(len(uniqueword))
```
