```python
import numpy as np
arr1 = np.array([1,2,3,4,5])
print(arr1.ndim)
```

```python
arr2 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(arr2.ndim)
```

```python
arr3 = np.array([1,2,3,4,5],ndmin=10)
print(arr3)
print(arr3.ndim)
```
