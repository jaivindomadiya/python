<h4>Constructors</h4>

```python
## Constructor is a INIT function
## it is intoke when object is created
class student:
    def __init__(self):
        pass ## this is default constructor
        ## there are two type of constructors 1)Default Constructors and 2) Parameterized Constructors
    
    def __init__(self,fullname): ##Constructor is always take one argument which is self so you need to pass self keyword in this constructor
        self.name = fullname
        print("Adding Value of New Student")
  
s1 = student("Jaivin")
print(s1.name)
```

```python
#Create a student class and take name and marks of three subjects and print the average of this three subjects
```

```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    def getAvg(self):
        sum = 0
        for value in self.marks:
            sum += value
        print("Hii",self.name,"Your Average is = ",sum/3)
s1 = Student("Jaivin",[98,98,88])
s1.getAvg()

s1.name = "Ironman" # so we can change value of attribute at runtime also okay 
s1.getAvg()
```

```python
##Static Methods
##Static methods is a method which don't have self keyword
class Student:
    def __init__(self):
        pass
    def hello():
        print("Hello ")

p1 = Student()
p1.hello()  ##it Gives me Error bcz by default hello ke under self hona chahiye okay 
```

```python
##Static Methods
##Static methods work at class level
class Student:
    def __init__(self):
        pass
    @staticmethod  ##decorator 
    def hello():
        print("Hello ")

p1 = Student()
p1.hello()   
```
