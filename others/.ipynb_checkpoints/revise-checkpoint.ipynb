```python
#Create a list of the first 20 positive integers. Print the list.
lst = list(range(1,21))
print(lst)
```

```python
#Print the first, middle, and last elements of the list created in Assignment 1.
print(lst[0])
print(lst[-1])
print(lst[len(lst) // 2])  ## for middle Element
```

```python
lst = [x**2 for x in range(1,21)]
print(lst)
```

```python
lst = [x for x in range(1,21) if x%2==0]
print(lst)
```

```python
def intersection(lst1,lst2):
    intersection = [x for x in lst1 if x in lst2]
    return intersection

lst1 = [1,2,3,4,5,6,7]
lst2 = [3,4,5,6,7]
print(intersection(lst1,lst2))
```

<h2>Lambda Function</h2>

```python
addition = lambda a,b:a+b
print(addition(5,6))
```

```python
even = lambda num:num%2==0
even(2)
```

```python
numbers = [2,3,4,5,6]
list(map(lambda x:x**2,numbers))
```

```python
list(map(lambda x:x%2==0,numbers))
```

<h2>Filter Function</h2>

```python
#filter(first parameter is function, second parameter is list)
#so syntex is : filter(function,list)
#Example:
lst = [1,2,3,4,5,6,7,8,9]
def even(num):
    if num%2==0:
        return num

list(filter(even,lst))
```
