```python
data = pd.read_excel("ESD.xlsx")
print(data)
```

```python
print(data.head(10))
```

```python
gp = data.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gp)
```

```python
gp1 = data.groupby("Country").agg({"Age":"mean"})
print(gp1)
```

```python
gp1 = data.groupby("Country").agg({"Annual Salary":"max"})
print(gp1)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max"})
print(gp1)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"min"})
print(gp1)
```

```python
import pandas as pd
data1 = {
    "EmpID": ["E1","E2","E3","E4","E5","E6"],
    "Name" : ["Ram","Shyam","Rajni","Vishal","John","Raj"],
    "Age" : [34,56,25,33,34,34]}

data2 = {
    "EmpID": ["E1","E2","E3","E4","E5","E6"],
    "Salary" : [45000,18000,25000,30000,24000,19000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2, on = "EmpID")) ## merge only common value okay
```

```python
print(pd.merge(df1,df2, on = "EmpID",how = "left")) ## Left Join 
```
