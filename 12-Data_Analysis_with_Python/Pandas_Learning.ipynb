```python
## Mearging and Joining Data Frame

df1 = pd.DataFrame({'Key': ['A','B','C'],'Value1': [1,2,3]})
df2 = pd.DataFrame({'Key': ['D','E','F'],'Value2': [4,5,6]})
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```
