**Question 3:** Write a Python program to check if a number is positive, negative, or zero.

```python
Number = int(input("Enter Number : "))
if Number > 0:
    print("Number is Positive")
elif Number < 0:
    print("Number is Negative")
else:
    print("Zero")
```

**Question 4:** Write a Python program to find the largest of three numbers.

```python
num1 = int(input("Enter Number 1 : "))
num2 = int(input("Enter Number 2 : "))
num3 = int(input("Enter Number 3 : "))

if num1 > num2 and num1 > num3:
    print("Number 1 ",num1,"is Largest from all three numbers")
elif num2 > num1 and num2 > num3:
    print("Number 2 ",num2,"is Largest from all three numbers")
elif num3 > num2 and num3 > num1:
    print("Number 3 ",num3,"is Largest from all three numbers")
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
number = int(input("Enter Aay Number :"))
fact = 1
for i in range(number):
    fact = number * fact
    number = number - 1
print(fact)
    
```

## 3. Basic Operators (Arithmetic, Comparison, Logical)

**Question 10:** Write a Python program to demonstrate comparison operators: equal to, not equal to, greater than, less than.

```python
a = 10
b = 20

print(a==b)
print(a!=b)
print(a>b)
print(a<b)
```

**Question 11:** Write a Python program to demonstrate logical operators: and, or, not.

```python
a = 10
b = 20
print(a and b) #b
print(a or b) #a
print(not a) #false
```

**Question 12** Write a Python program to calculate the square of a number.

```python
number = int(input("Enter Num:"))
squre = number * number
print(squre)
```

**Question 13:** Write a Python program to find the sum of the first n natural numbers.

```python
n = int(input("Enter Number:"))
sum = 0
for i in range(1,n):
    sum = sum + i
print(sum)
```
