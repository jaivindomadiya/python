```python
print(f"first Element : {tpl[0]}") 
print(f"last Element : {tpl[9]}") 
print(f"Middle Element : {tpl[len(tpl)//2]}") 

```

<h4>Assignment 3: Tuple Slicing ::
Print the first three elements, the last three elements, and the elements from index 2 to 5 of the tuple created in Assignment 1.</h4>

```python
print(tpl[0:3]) #First Three Element
print(tpl[-3:]) #last three element
print(tpl[2:5]) #2 to 5
```
