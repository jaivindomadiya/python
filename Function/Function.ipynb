```python
## Positional Arguments

def printnumbers(*args):
    for i in args:
        print(i,end=" ")
printnumbers(1,2,3,4,5,6,67,7,8)
```

```python
##Keyword Arguments

def printkeyvaluepairs(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
printkeyvaluepairs(name="Jai",age=22,city="Junagadh")
```

```python
def printkeyvaluepairs(*args,**kwargs):
    for val in args:
        print(val,end=" ")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
printkeyvaluepairs(name="Jai",age=22,city="Junagadh")
printkeyvaluepairs(1,2,3,4,5)
```

```python
##1. Temprature Conversion
def temprature(temp,con):
    if con == 'c':
        return temp * 9/5 + 32 #celcius to faranhit
    elif con == 'f':
        return (temp-32)*5/9 #faranhit to celcius
    else:
        return None

print(temprature(25,'c'))
```

```python
##Pasward Strength Checker
def passwordchk(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    return True
```

```python
passwordchk("P@tEl@123#")
```

```python
passwordchk("jaivin")
```

```python
#password chacker

def is_strongpassword(password):
    #for checking password is strong or not
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()+-' for char in password):
        return False
    return True
```

```python
is_strongpassword("Jaivin@2002")
```
