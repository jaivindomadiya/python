```python
sum = 0
n = 1
while n != 0:
    if n == 0:
        break
    else:
        n = int(input("Enter Number: "))
        sum = sum + n   
print(sum)
```

```python
# Factorial Calculation
fact = 1
no = int(input("Enter Number:"))
for i in range(no):
    fact = fact * no
    no = no-1
print(fact)
```
