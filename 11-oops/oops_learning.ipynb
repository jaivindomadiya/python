<h2>Inheritance</h2>

```python
class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"Person will Driving {self.enginetype} Car")


car1 = Car(4,4,"petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving = is_selfdriving

    def selfdrive(self):
        print(f"Tesla Supports Self Driving Car{self.is_selfdriving}")

tesla1 = Tesla(4,5,"Electric",True)
tesla1.selfdrive()
        
#It is Single Inheritance okay
```

```python
#Multiple Inheritance
#When a class inherits from more than one base class .

#Base calss 1
class Animal:
    def __init__(self,name):
        self.name = name
    def speak(self):
        print(f"Sub class must implement  this method")

#Base calss 2
class Pet:
    def __init__(self,owner):
        self.owner = owner
    def method1(self):
        print(f"pet method 1")

## Derived calss
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} say wolf"

##Create an object
dog = Dog("Buddy","Jaivin")

print(dog.speak())
print(f"owner : {dog.owner}")
```

<h2>Polymorphism</h2>

```python
#method overriding
#it allows a child class to priovide a specific implementation of a method that is already define in its parent class 
```

```python
class Animal:
    def speak(self):
        return "Sound speak animal"

#Derived class 1
class Dog(Animal):
    def speak(self):
        return "Bauuu"
    

#Derived class 2
class Cat(Animal):
    def speak(self):
        return "It Rerurns meow"


#Function that demostrate polymorphism
def animal_speak(animal):
    print(animal.speak())


dog = Dog()
print(dog.speak())

animal1 = Animal()
print(animal1.speak())

animal_speak(dog) #function 

cat = Cat()
print(cat.speak())
        
```

```python
#Polymorphism with functions and methods 
#Base class

class Shape:
    def area(self):
        return "The Area of the Figure"

#Derived Calss 1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

#Derived Calss 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return (self.radius * self.radius) * 3.14

#Funcations that demostrate polymorphism
def print_area(shape):
    print(f"This is {shape.area()}")


rect = Rectangle(4,5)
cir = Circle(3)

print_area(rect)
print_area(cir)

```
